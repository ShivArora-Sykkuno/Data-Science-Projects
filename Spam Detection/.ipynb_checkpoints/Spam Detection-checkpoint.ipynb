```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("data/SMSSpamCollection", names=["label", "messages"], sep="\t")
data.head()
```

```python
import nltk
nltk.download('punkt')
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

wl = WordNetLemmatizer()
corpus = []
for i in range(len(data)):
    sentence = re.sub('[^a-zA-Z]', ' ', data["messages"][i])
    sentence = sentence.lower()
    sentence = sentence.split()
    sentence = [wl.lemmatize(word) for word in sentence if not word in set(stopwords.words("english"))]
    sentence = ' '.join(sentence)
    corpus.append(sentence)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

td = TfidfVectorizer()
X = td.fit_transform(corpus).toarray()
```

```python
from sklearn.preprocessing import LabelEncoder

l1 = LabelEncoder()
Y = l1.fit_transform(data["label"])
```

```python
# l1.inverse_transform(Y) -> For  Reversible Transformation or Decoding the Encoding
Y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=101)
```

```python
from sklearn.ensemble import RandomForestClassifier

spam_detection = RandomForestClassifier()
spam_detection.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix

y_preds = spam_detection.predict(X_test)
confusion_matrix(y_test, y_preds)
```

```python
from sklearn.metrics import plot_confusion_matrix

plot_confusion_matrix(spam_detection, X_test, y_test)
```

```python
import joblib

joblib.dump(spam_detection, 'model/spam_detection.pkl')
```
